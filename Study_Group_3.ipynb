```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = '/root/.cache/kagglehub/datasets/ankushpanday1/global-road-accidents-dataset/versions/1/road_accident_dataset.csv'
df = pd.read_csv(url)

```

```python
# visualisasi distribusi Tingkat Keparahan Kecelakaan
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Accident Severity', palette='cool')
plt.title('Tingkat Keparahan Kecelakaan')
plt.xlabel('Severity')
plt.ylabel('Count')
plt.show()

# Visualisasi jumlah kendaraan yang terlibat
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Number of Vehicles Involved', bins=4, kde=False, color='skyblue')
plt.title('jumlah kendaraan yang terlibat')
plt.xlabel('Number of Vehicles')
plt.ylabel('Count')
plt.show()

# visualisasi penyebab kecelakaan
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='Accident Cause', order=df['Accident Cause'].value_counts().index, palette='viridis')
plt.title('penyebab kecelakaan')
plt.xlabel('Count')
plt.ylabel('Accident Cause')
plt.show()
```

Dataset ini terdiri dari Negara tempat kecelakaan terjadi,
Informasi waktu kecelakaan,
Waktu kejadian (Pagi, Siang, dll.),
Apakah kejadian terjadi di area urban atau rural,
Tingkat keparahan kecelakaan,
umlah cedera dan kematian,
Kondisi jalan saat kecelakaan,
Penyebab kecelakaan

Dari visualisasi awal:

Tingkat keparahan kecelakaan :
Distribusi menunjukkan mayoritas kecelakaan masuk dalam kategori tingkat menengah atau parah.

Jumlah kendaraan yang terlibat :
Kecelakaan dengan dua atau tiga kendaraan paling sering terjadi, sementara kecelakaan yang melibatkan satu atau empat kendaraan lebih jarang.

Penyebab kecelakaan :
Penyebab paling umum adalah "Distracted Driving" (mengemudi dalam kondisi terganggu), diikuti oleh "Speeding" (kecepatan tinggi) dan "Weather" (kondisi cuaca buruk).
